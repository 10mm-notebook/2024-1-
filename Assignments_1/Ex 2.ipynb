```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('practice1.txt', nodetype=str, data=[('label', str)])

# Draw the graph
nx.draw(G, pos=nx.shell_layout(G), with_labels=True)
plt.show()
```

```python
#Print the the total number of nodes and edges
n=G.number_of_nodes()
print(n)
m=G.number_of_edges()
print(m)
```

```python
#Print a list of all nodes
print(list(G.nodes()))
```

```python
#Print all the neighbours of node David
print(list(G.neighbors('David')))
```

```python
#Print all the neighbours of node David who are its friends
for u, v, data in G.edges('David', data=True):
    if data['label'] == 'friend':
        print(v)
```

```python
#What is the degree of node David in the graph?
print(G.degree('David'))
```

```python
#What is the relation type between David and Amy?
print(G['David']['Amy']['label'])
```

```python
#Is the graph a connected graph?
is_connected = nx.is_connected(G)
if is_connected:
    print("The graph is a connected graph.")
else:
    print("The graph is a disconnected graph.")
```

```python
#What is the average degree of the nodes in the graph?

degrees = G.degree()
print(sum(dict(degrees).values())/ len(G.nodes()))
```

```python
#Which nodes in the graph have a degree greater than 3?
for node in (G.nodes()):
  if G.degree(node)>3:
    print(node)
```
