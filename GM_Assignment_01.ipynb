```python
import networkx as nx
import networkx
import matplotlib.pyplot as plt
```

```python
#add a node
G= networkx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)
G.add_node(7)
G.add_node(8)
G.add_node(9)

#add an edge
G.add_edge(1, 2)
G.add_edge(1, 6)
G.add_edge(2, 3)
G.add_edge(2, 4)
G.add_edge(2, 6)
G.add_edge(3, 4)
G.add_edge(3, 5)
G.add_edge(4, 8)
G.add_edge(4, 9)
G.add_edge(6, 7)
G.add_edge(8, 9)
```

```python
#1.Draw the given graph
nx.draw(G,with_labels=True)
```

```python
#2.Print the total number of nodes in the graph
n=G.number_of_nodes()
print(n)

#3.Print the total number of edges in the graph
m=G.number_of_edges()
print(m)
```

```python
#4.Provide a list of all edges in the graph
edge_list=G.edges()
print(edge_list)
```

```python
#5.Find degree of a node 1
print(G.degree(1))
```

```python
#6.Determine if the given graph is connected or if there are disconnected components. If disconnected, identify and describe each component
if nx.is_connected(G):
    print("the given graph is connected")
else:
    print("the given graph is disconnected.")
    for component in nx.connected_components(G):
        print(component)
```

```python
#7.Calculate and display the degree distribution for the entire graph.
degrees = [G.degree(n) for n in G.nodes()]


degree_distribution = {}
for degree in degrees:
    if degree in degree_distribution:
        degree_distribution[degree] += 1
    else:
        degree_distribution[degree] = 1


print("degree distribution")
for degree, count in degree_distribution.items():
    print(f"degree {degree}: {count} number of nodes")


plt.bar(degree_distribution.keys(), degree_distribution.values())
plt.xlabel('Degree')
plt.ylabel('Number of Nodes')
plt.title('Degree Distribution of the Graph')
plt.show()
```
